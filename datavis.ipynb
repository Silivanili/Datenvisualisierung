```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


df = df.dropna(subset=['Publishers', 'Average playtime forever'])


avg_plt_per_publ = (
    df.groupby('Publishers')['Average playtime forever']
      .mean()
      .sort_values(ascending=False)          
      .head(10)                               
)


norm = plt.Normalize(vmin=avg_plt_per_publ.min(),
                     vmax=avg_plt_per_publ.max())

cmap = plt.cm.Blues

colours = cmap(norm(avg_plt_per_publ.values))


ax = avg_plt_per_publ.plot(
    kind='barh',
    figsize=(10, 8),
    color=colours,          
    edgecolor='black'       
)

ax.set_xlabel('Average playtime (forever)')
ax.set_title('Average playtime forever per publisher (top 10)')
plt.tight_layout()
plt.show()
```
